```python
import pickle as pkl
import shutil
import os

with open("reactants_products_sep1a_filtered_dft_opted_results.pkl", "rb") as f:
    results = pkl.load(f)

for failed_dict, success_dict in results:
    if failed_dict:
        for mol_id in failed_dict:
            ids = str(int(int(mol_id.split("id")[1]) / 1000))
            if os.path.exists(
                os.path.join(
                    "output",
                    "FF_conf",
                    "outputs",
                    f"outputs_{ids}",
                    f"{mol_id}_confs.sdf",
                )
            ):
                #                 pass
                os.remove(
                    os.path.join(
                        "output",
                        "FF_conf",
                        "outputs",
                        f"outputs_{ids}",
                        f"{mol_id}_confs.sdf",
                    )
                )
            if os.path.exists(
                os.path.join(
                    "output",
                    "semiempirical_opt",
                    "outputs",
                    f"outputs_{ids}",
                    f"{mol_id}.tar",
                )
            ):
                #                 pass
                os.remove(
                    os.path.join(
                        "output",
                        "semiempirical_opt",
                        "outputs",
                        f"outputs_{ids}",
                        f"{mol_id}.tar",
                    )
                )
            if os.path.exists(
                os.path.join(
                    "output",
                    "DFT_opt_freq",
                    "outputs",
                    f"outputs_{ids}",
                    f"{mol_id}.log",
                )
            ):
                #                 pass
                os.remove(
                    os.path.join(
                        "output",
                        "DFT_opt_freq",
                        "outputs",
                        f"outputs_{ids}",
                        f"{mol_id}.log",
                    )
                )
            if os.path.exists(
                os.path.join(
                    "output", "DLPNO_sp", "outputs", f"outputs_{ids}", f"{mol_id}.log"
                )
            ):
                #                 pass
                os.remove(
                    os.path.join(
                        "output",
                        "DLPNO_sp",
                        "outputs",
                        f"outputs_{ids}",
                        f"{mol_id}.log",
                    )
                )
            if os.path.exists(
                os.path.join(
                    "output", "DLPNO_sp", "inputs", f"inputs_{ids}", f"{mol_id}.in"
                )
            ):
                #                 pass
                os.remove(
                    os.path.join(
                        "output", "DLPNO_sp", "inputs", f"inputs_{ids}", f"{mol_id}.in"
                    )
                )
            if os.path.exists(
                os.path.join(
                    "output", "COSMO_calc", "outputs", f"outputs_{ids}", f"{mol_id}.tar"
                )
            ):
                #                 pass
                os.remove(
                    os.path.join(
                        "output",
                        "COSMO_calc",
                        "outputs",
                        f"outputs_{ids}",
                        f"{mol_id}.tar",
                    )
                )
            if os.path.exists(
                os.path.join(
                    "output", "COSMO_calc", "outputs", f"outputs_{ids}", mol_id
                )
            ):
                #                 pass
                shutil.rmtree(
                    os.path.join(
                        "output", "COSMO_calc", "outputs", f"outputs_{ids}", mol_id
                    )
                )


sum(1 for failed_dict, success_dict in results if failed_dict)
```
